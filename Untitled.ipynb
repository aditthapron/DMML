```python
import numpy as np
import glob
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
file = np.load('out.npz')
```

```python
ind=file['ind']
gallery=file['gallery']
query=file['query']
```

```python
ind.shape
```

```python
gallery[0]
```

```python
print(ind[0:5][0:5])
```

```python
fig, ax = plt.subplots(nrows=10, ncols=11,figsize=(20,15))
for i in range(0,10):
    file = glob.glob('../VRIC/probe_images/'+str(query[i])+'_*.*')
    img=mpimg.imread(file[0])
    ax[i][0].imshow(img)
    ax[i][0].set_xticks([])
    ax[i][0].set_yticks([])
    for j in range(1,11):
        if ind[i][j]==i:
            ax[i][j].spines['bottom'].set_color('red')
            ax[i][j].spines['top'].set_color('red')
            ax[i][j].spines['left'].set_color('red')
            ax[i][j].spines['right'].set_color('red')
            ax[i][j].spines['bottom'].set_linewidth(7)
            ax[i][j].spines['top'].set_linewidth(7)
            ax[i][j].spines['left'].set_linewidth(7)
            ax[i][j].spines['right'].set_linewidth(7)
        file = glob.glob('../VRIC/gallery_images/'+str(gallery[ind[i][j]])+'_*.*')
        img=mpimg.imread(file[0])
        ax[i][j].imshow(img)
        ax[i][j].set_xticks([])
        ax[i][j].set_yticks([])
plt.savefig('all.jpg')
```

```python
temp=[]
for i in range(len(ind)):
    temp.append(np.argwhere(ind[i]==i)[0][0])
top_worst = np.argsort(temp)[::-1]
```

```python
fig, ax = plt.subplots(nrows=10, ncols=11,figsize=(20,15))
for i in range(0,10):
    index=ind[top_worst[i]]
    file = glob.glob('../VRIC/probe_images/'+str(query[top_worst[i]])+'_*.*')
    img=mpimg.imread(file[0])
    ax[i][0].imshow(img)
    ax[i][0].set_xticks([])
    ax[i][0].set_yticks([])
    for j in range(1,11):
        file = glob.glob('../VRIC/gallery_images/'+str(gallery[index][j])+'_*.*')
        img=mpimg.imread(file[0])
        ax[i][j].imshow(img)
        ax[i][j].set_xticks([])
        ax[i][j].set_yticks([])
plt.savefig('top_worst.jpg')
```
